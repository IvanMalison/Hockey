```python
from pylab import *
%matplotlib inline
import seaborn
import PbPMethods2 as pm2
```

```python
current = {'Backstrom': [9, 22, 38], #g, a, gp
           'Ovechkin': [18, 12, 38],
           'Johansson': [14, 14, 38],
           'Kuznetsov': [4, 21, 38],
           'Carlson': [5, 17, 38],
           'Williams': [11, 9, 38],
           'Oshie': [12, 7, 31],
           'Orlov': [2, 16, 38],
           'Niskanen': [2, 16, 37],
           'Beagle': [6, 9, 38],
           'Burakovsky': [5, 10, 35],
           'Winnik': [6, 4, 28],
           'Orpik': [0, 9, 38],
           'Schmidt': [1, 7, 31],
           'Eller': [3, 4, 37],
           'Alzner': [3, 3, 38],
           'Connolly': [4, 1, 24],
           'Wilson': [2, 2, 38]}
gr = 82 - 38
fs = ['Backstrom', 'Ovechkin', 'Johansson', 'Kuznetsov', 'Oshie', 'Williams',
        'Burakovsky', 'Beagle', 'Winnik', 'Eller', 'Connolly', 'Wilson']
ds = ['Carlson', 'Orlov', 'Niskanen', 'Orpik', 'Schmidt', 'Alzner']
proj_g = [current[x][0] + round(current[x][0]/current[x][2]*gr) for x in fs] + \
            [current[x][0] + round(current[x][0]/current[x][2]*gr) for x in ds]
proj_a = [current[x][1] + round(current[x][1]/current[x][2]*gr) for x in fs] + \
            [current[x][1] + round(current[x][1]/current[x][2]*gr) for x in ds]
i = [i for i in range(len(proj_g))]
```

```python
colors = seaborn.color_palette('muted')
barh(i, proj_g[::-1], color=colors[0], label='Goals pace')
barh(i, proj_a[::-1], left=proj_g[::-1], color=colors[1], label='Assists pace')
yticks([y + 0.4 for y in i], ds[::-1] + fs[::-1])
legend(loc=2, bbox_to_anchor=(1, 1), title='If they play all\nremaining\n{0:d} games...'.format(gr))
tick_params(labeltop=True)
```
